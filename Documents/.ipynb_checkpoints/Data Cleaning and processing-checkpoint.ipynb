# 2.1. Reviews cleaning

We will create a function to clean the review.

```python
#we will just apply this to our object columns, this removes the extra spaces. I create a part a function, because
#it can be useful to use it another time

def remove_extra_spaces(data): 
    for column in data.select_dtypes(include=["object"]).columns:
        data[column] = data[column].apply(lambda text: text.strip())
    return data
```

We will remove the non alpha- numeric characters, convert all to lower case, email adresses or html tags or websites.

```python
### I would like to see how many entries hae digits on the Review colum

kas[kas["Review"].str.contains(r'\d+')]["Review"] #466 reviews have numbers...
```

```python
def clean_reviews(data):
    
    for column in data.select_dtypes(include=["object"]).columns:
    
        #we convert all the word to lowe case
        data[column] = data[column].apply(lambda text: text.lower())
        # replace non alphanumeric by spaces
        data[column] = data[column].apply(lambda text : re.sub(r'\W+',' ',text))
        # Replace urls by ''
        data[column] = data[column].apply(lambda text : re.sub(r'([a-z0-9+._-]+@[a-z0-9+._-]+\.[a-z0-9+_-]+)',' ',text))
        #Remove sites
        data[column] = data[column].apply(lambda text : re.sub(r'www.\S+', ' ' ,text))
        #Remove Url's
        data[column] = data[column].apply(lambda text : re.sub(r'^(http|https|ftp|ssh)\S+', ' ',text))
        #Remove digits
        data[column] = data[column].apply(lambda text : re.sub(r'\d+',' ',text))
        #apply the extra spaces function defined before
        remove_extra_spaces(data)

    return data
   


## We need to think about numbers and euros....    
```

```python
kas = clean_reviews(kas)
```

## 2.2. Create the corpus

We will create a corpus, database, of words in the reviews. This is an specific set of common words in the reviews for restaurant, for example.

I will create a corpus for the header of the reviews and the body.

There is a function also for that: word_tokenize(string)

```python
###the words in the whole set of reviews
def create_corpus(data, column):
    """
    Input a data frame, and the a string of the name of the column you want to create the corpus
    """
 
    #create an empty string
    reviews = ''

    for review in data[column]:
        reviews += review
    corpus = nltk.sent_tokenize(reviews)
    
    return corpus
```

```python
corpus_body_reviews = create_corpus(kas,"Review")
corpus_body_reviews 
```

## 2.3.Tokenizing the corpus

We will create a dictionary that contains the words fo the corpus as keys and the amount of times that appreat as value. We will now the frecuence for each

```python
wordfreq = {}

for sentence in corpus_body_reviews:
    
    words = sentence.split()
    
    for word in words:
        if ( word not in wordfreq.keys() ):
            wordfreq[word] = 1 # We initialize the corresponding counter
        else:
            wordfreq[word] += 1 # We increase the corresponding counter
```

```python
len(list(wordfreq.keys()))
```

```python
wordfreq
```

## 2.4. Clean corpus

We can clean a little bit further our corpus. 

NLTK has a corpus (set of words) that can be use to clean out corpus like: pronouns, single letters...]


```python
from nltk.corpus import stopwords

stop_words = list(stopwords.words('english')) 
stop_words
```

```python
#we need to access by index in the list
for i in range(len(stop_words)):
    stop_words[i] = stop_words[i].replace("'"," ")
    
#i can also drop single letters
#stop_words = [word for word in list(np.unique(stop_words)) if len(word) > 1]
stop_words
```

We will retrieve the words that are not in the stop words

```python
corpus = [(wordfreq[key],key) for key in list(wordfreq.keys()) if key not in stop_words]
```
